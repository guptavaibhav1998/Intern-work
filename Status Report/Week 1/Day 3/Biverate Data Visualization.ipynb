```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
reviews=pd.read_csv('data.csv')
reviews.head(10)
```

```python
#Ploting price vs points of all data
reviews[reviews['price'] < 100].plot.scatter(x='price', y='points')
```

```python
#We can apply plot directly on our database i.e reviews
reviews.plot.scatter('price','points')
```

```python
#taking subset of data(30 entries)
reviews[reviews['price']<100].sample(30).plot.scatter(x='price',y='points')
```

```python
#Hexplot
reviews[reviews['price'] < 100].plot.hexbin(x='price', y='points',gridsize=20)
```
