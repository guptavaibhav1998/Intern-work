```python
#Seaborn
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd
import numpy as np
```

```python
#StripPlot(used for plotting categorical data) without jitter
data=sb.load_dataset('iris')
sb.stripplot(x='species',y='petal_length', data=data,jitter=False)
plt.show()
#Here scatter plots are overlapped so we add some random noise(jitter)
```

```python
#Stripplot with jitter
data=sb.load_dataset('iris')
sb.stripplot(x='species',y='petal_length',data=data,jitter=True)
plt.show()
```

```python
#Scatter PLot
data=pd.read_csv('pokemon.csv')
sb.lmplot(x='attack',y='defense',data=data.head(10))

#Alternate way
#sb.lmplot(x=data.attack,y=data.defense)
```

```python
#Scatter Plots without regression line
sb.lmplot(x='attack',
          y='defense',
          data=data,
          fit_reg=False,#No regression line
          hue='generation')#color on the basis of generation
```

```python
#Limits integration with matplotlib
#plt.xlim(l,r)  this limit the values on x axis
#that is it will show values from l to r on x-axis
sb.lmplot(x='attack',
         y='defense',
         data=data,
         fit_reg=False,
         hue='generation')

plt.ylim(50,150)
plt.xlim(20,100)
```

```python
#Box plot
plt.figure(figsize=(10,6))

sb.boxplot(x='type1',y='defense',data=data)

```

```python
#Violin Plot
sb.violinplot(x='type1',y='attack',data=data)

plt.xlim(0,10)
```

```python
#Violin plot with color palette
plt.figure(figsize=(10,6))
colors = ['#78C850',  # Grass
                    '#F08030',  # Fire
                    '#6890F0',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#7038F8',  # Dragon
                   ]

sb.violinplot(x='type1',y='attack',data=data,palette=colors)
```

```python
#Swarm Plot
plt.figure(figsize=(10,6))
sb.swarmplot(x='type1',y='attack',data=data)
```

```python
#Overlapping Plots
plt.figure(figsize=(10,6))
 
# Create plot
sb.violinplot(x='type1',
              y='attack', 
              data=data,
              inner=None #Remove the bars inside the violin     
             )
 
sb.swarmplot(x='type1', 
             y='attack', 
             data=data,
             color='black', #Makes point black
             alpha=0.5      #Makes point slightly transparent
            )
 
```

```python
#Heat Map

# Calculate correlations
corr = data.corr()
 
# Heatmap
sb.heatmap(corr)
```

```python
#Histogram

sb.distplot(data.attack)
```

```python
#Count Plot (a.k.a. Bar Plot)
sb.countplot(x='type1', data=data)
 
# Rotate x-labels
plt.xticks(rotation=-45)
```

```python
# Factor Plot
#Factor plots make it easy to separate plots by categorical classes.

g = sb.factorplot(x='type1', 
                  y='attack', 
                  data=data, 
                  hue='generation',  # Color by generation
                  col='generation',  # Separate by generation
                  kind='swarm') # Swarmplot
 
# Rotate x-axis labels
g.set_xticklabels(rotation=-45)
 
# Doesn't work because only rotates last plot
# plt.xticks(rotation=-45)
```

```python
# Density Plot(KDEPLOT)
sb.kdeplot(data.attack, data.defense)
```

```python
#KdePlot with Scatter Plot
sb.kdeplot(data.attack, data.defense)
sb.scatterplot(data.attack,data.defense)
#So kdePlot tells in which particular area density is highest
```
